```python
%matplotlib inline
import csv
import numpy
import matplotlib.pyplot as plt
```

```python
rgb_prec =  open('./rgb_precision.csv', 'r')
rgb_recall =  open('./rgb_recall.csv', 'r')

rgbd_1_prec =  open('./rgbd-1_precision.csv', 'r')
rgbd_1_recall =  open('./rgbd-1_recall.csv', 'r')

rgbd_2_prec =  open('./rgbd-2_precision.csv', 'r')
rgbd_2_recall =  open('./rgbd-2_recall.csv', 'r')

rgbd_3_prec =  open('./rgbd-3_precision.csv', 'r')
rgbd_3_recall =  open('./rgbd-3_recall.csv', 'r')
```

```python
f_rgb_prec = csv.reader(rgb_prec)
f_rgb_recall = csv.reader(rgb_recall)

f_rgbd_1_prec = csv.reader(rgbd_1_prec)
f_rgbd_1_recall = csv.reader(rgbd_1_recall)

f_rgbd_2_prec = csv.reader(rgbd_2_prec)
f_rgbd_2_recall = csv.reader(rgbd_2_recall)

f_rgbd_3_prec = csv.reader(rgbd_3_prec)
f_rgbd_3_recall = csv.reader(rgbd_3_recall)
```

```python
rgb_prec_ary = []
rgb_recall_ary = []

rgbd_1_prec_ary = []
rgbd_1_recall_ary = []

rgbd_2_prec_ary = []
rgbd_2_recall_ary = []

rgbd_3_prec_ary = []
rgbd_3_recall_ary = []

```

```python
for prec, recall in zip(f_rgb_prec, f_rgb_recall):
    rgb_prec_ary.append(prec)
    rgb_recall_ary.append(recall)
    
for prec, recall in zip(f_rgbd_1_prec, f_rgbd_1_recall):
    rgbd_1_prec_ary.append(prec)
    rgbd_1_recall_ary.append(recall)
    
for prec, recall in zip(f_rgbd_2_prec, f_rgbd_2_recall):
    rgbd_2_prec_ary.append(prec)
    rgbd_2_recall_ary.append(recall)

for prec, recall in zip(f_rgbd_3_prec, f_rgbd_3_recall):
    rgbd_3_prec_ary.append(prec)
    rgbd_3_recall_ary.append(recall)
```

```python
plt.figure()
plt.plot(rgb_recall_ary[13], rgb_prec_ary[13])
plt.plot(rgbd_1_recall_ary[13], rgbd_1_prec_ary[13])
plt.plot(rgbd_2_recall_ary[13], rgbd_2_prec_ary[13])
plt.plot(rgbd_3_recall_ary[13], rgbd_3_prec_ary[13])
plt.legend(['SSD', 'Model C', 'Model A', 'Model B'])
```
